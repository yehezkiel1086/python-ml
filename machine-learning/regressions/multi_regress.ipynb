# Multiple Linear Regression

Multiple regression is like linear regression, but with more than one independent value, meaning that we try to predict a value based on two or more variables. (W3Schools)

Case:
- Find the Multiple Linear Regression model of [cars_dataset.csv](cars_dataset.csv)
- predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3
- Find the coeficients
- Visualize the models and datas

```python
import pandas

# import the csv data
df = pandas.read_csv("cars_dataset.csv")

df.head()
```

```python
from sklearn import linear_model

# create the multi regress model attributes
X = df[["Weight", "Volume"]]
Y = df["CO2"]

# create the multi regress model
regress = linear_model.LinearRegression()
regress.fit(X, Y)
```

```python
# predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3

predicted_args = [2300, 1300]

predicted_res = regress.predict([predicted_args])

predicted_res
```

```python
# find the coefficients

regress.coef_
```
