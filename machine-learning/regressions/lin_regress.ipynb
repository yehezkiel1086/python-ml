# Linear Regression

Linear regression uses the relationship between the data-points to draw a straight line through all them. (W3Schools)

In statistics, linear regression is a model that estimates the relationship between a scalar response and one or more explanatory variables. A model with exactly one explanatory variable is a simple linear regression; a model with two or more explanatory variables is a multiple linear regression. (Wikipedia)

Find linear regression model of the following dataset and visualize it:

```python
X = [5,7,8,7,2,17,2,9,4,11,12,9,6]
Y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(X, Y)

def calc_y(x):
  return x * slope + intercept

Y_model = list(map(calc_y, X))
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, Y)
plt.plot(X, Y_model)
plt.show()
```

Input a data then try predicting the output, also visualize it!

```python
x_predict = int(input("Enter data to predict: "))
y_predict = calc_y(x_predict)

print(f"Prediction outcome = {y_predict}")
print(f"R value = {r}")

plt.scatter(X, Y)
plt.plot(X, Y_model)
plt.axhline(y_predict, color="orange", linewidth=1)
plt.axvline(x_predict, color="orange", linewidth=1)
plt.show()
```
