# Polynomial Regression

If your data points clearly will not fit a linear regression (a straight line through all data points), it might be ideal for polynomial regression.

Polynomial regression, like linear regression, uses the relationship between the variables x and y to find the best way to draw a line through the data points. (W3Schools)

Case:

- From the following data:
    ```py
    x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
    y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
    ```

    1. Create the polynomial regression model
    2. Visualize the data
    3. Input and predict a value, then visualize it
    4. Find the R2 score

````python
import matplotlib.pyplot as plt
import numpy

X = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
Y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

# create the polynomial regression model
Y_model = numpy.poly1d(numpy.polyfit(X, Y, 3))
X_model = numpy.linspace(1, 22, 100)

# Visualize the model data
plt.scatter(X, Y)
plt.plot(X_model, Y_model(X_model))
plt.show()
````

````python
# Input and predict a value, then visualize it

x_predict = int(input("Enter a value to predict: "))
y_predict = Y_model(x_predict)

print(f"Prediction result: {y_predict}")

# Prediction visualization
plt.scatter(X, Y)
plt.plot(X_model, Y_model(X_model))
plt.axvline(x=x_predict, color="orange", linewidth=1)
plt.axhline(y=y_predict, color="orange", linewidth=1)
plt.show()
````

````python
# find the r2 score

from sklearn.metrics import r2_score

r2_result = r2_score(Y, Y_model(X))

print(f"R2 score: {r2_result}")
````
