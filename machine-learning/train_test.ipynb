## Train / Test

Train/Test is a method to measure the accuracy of your model.

It is called Train/Test because you split the data set into two sets: a training set and a testing set. (W3Schools)

> 80% for training, and 20% for testing.

### Case

Our data set illustrates 100 customers in a shop, and their shopping habits.

Split / Train the Dataset!

Define the datas:

```python
import numpy
import matplotlib.pyplot as plt
numpy.random.seed(2)

X = numpy.random.normal(3, 1, 100)
Y = numpy.random.normal(150, 40, 100) / X

plt.scatter(X, Y)
plt.show() 
```

Split the train and test datas:

```python
X_train = X[:80]
Y_train = Y[:80]

X_test = X[80:]
Y_test = Y[80:]

plt.scatter(X_train, Y_train)
plt.show()
```

```python
plt.scatter(X_test, Y_test)
plt.show()
```

Define the polynomial model:

```python
import numpy

Y_model = numpy.poly1d(numpy.polyfit(X_train, Y_train, 4))
model_line = numpy.linspace(0, 6, 100)

plt.scatter(X_train, Y_train)
plt.plot(model_line, Y_model(model_line))
plt.show()
```

Find out the R2 score:

```python
from sklearn.metrics import r2_score

train_r2 = r2_score(Y_train, Y_model(X_train))

train_r2
```

Bring in the test dataset:

```python
r2_test = r2_score(Y_test, Y_model(X_test))

r2_test
```

Try doing prediction on user input / random value:

```python
x_predict = int(input("Enter a value to predict: "))
y_predict = Y_model(x_predict)

print(f"Value to predict: {x_predict}; predicted value: {y_predict}")

plt.scatter(X, Y)
plt.plot(model_line, Y_model(model_line))
plt.axvline(x_predict, color="orange", linewidth=1)
plt.axhline(y_predict, color="orange", linewidth=1)
plt.show()
```
