# Data Distribution

2 types of data distributions:
1. Uniform data distribution
2. Normal data distribution

## Uniform Data Distribution
Cases:
1. Create an array containing 250 random floats between 0 and 5, then visualize it! (divide the visualization data by 5)
2. Create an array containing 100000 random floats between 0 and 5, then visualize it! (divide the visualization data by 100)

```python
import numpy

# case 1: 250 random floats between 0 and 5
arr_1 = numpy.random.uniform(0.0, 5.0, 250)

arr_1
```

```python
import matplotlib.pyplot as plt

plt.hist(arr_1, 5)
plt.show()
```

```python
# case 2: 100000 random floats between 0 and 5
arr_2 = numpy.random.uniform(0.0, 5.0, 100000)
plt.hist(arr_2, 100)
plt.show()
```

## Normal Data Distribution

Normal data distribution is an array of values where the datas are concentrated around a given value

In probability theory this kind of data distribution is known as the normal data distribution, or the Gaussian data distribution, after the mathematician Carl Friedrich Gauss who came up with the formula of this data distribution. (W3Schools)

Case:

Create 100000 normal data distribution with center of 5.0 and standard deviation of 1.0, then visualize it

```python
arr_3 = numpy.random.normal(5.0, 1.0, 100000)

plt.hist(arr_3, 100)
plt.show()
```
